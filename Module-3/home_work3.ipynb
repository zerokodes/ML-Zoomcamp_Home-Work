```python
from sklearn.feature_extraction import DictVectorizer
```

```python
dicts = df_categorical_columns[['y']].iloc[:100].to_dict(orient='records')
dicts
```

```python
dv = DictVectorizer(sparse=False)
```

```python
dv.fit(dicts)
```

```python
dv.transform(dicts)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df_categorical_columns, test_size=0.2, random_state=42)
```

```python
len(df_full_train), len(df_test)
```

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.y.values
y_val = df_val.y.values
y_test = df_test.y.values
```

```python
del df_train['y']
del df_val['y']
del df_test['y']
```

```python
from sklearn.metrics import mutual_info_score
```

```python
def mutual_info_y_score(series):
    return mutual_info_score(series, df_full_train.y)
```

```python
mi = df_full_train[categorical_columns].apply(mutual_info_y_score).round(2)
mi.sort_values(ascending=False)
```

```python
categorical = ['age','job','marital','education','balance','housing','contact','day','month','duration','campaign','pdays','previous','poutcome']
categorical
```

```python
train_dicts = df_train[categorical].to_dict(orient='records')
train_dicts
```

```python
X_train = dv.fit_transform(train_dicts)
```

```python
val_dicts = df_val[categorical].to_dict(orient='records')
```

```python
X_val = dv.transform(val_dicts)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
```

```python
model.fit(X_val, y_val)
```

```python
y_pred = model.predict_proba(X_val)[:,1]
y_pred
```

```python
y_decision = (y_pred >= 0.5)
y_decision
```

```python
accuracy_score =( y_val == y_decision).mean()
accuracy_score
```
